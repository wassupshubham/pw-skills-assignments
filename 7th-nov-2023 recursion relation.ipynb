To find the value of \( T(2) \) for the recurrence relation \( T(n) = 3T(n-1) + 12n \) with \( T(0) = 5 \), we will compute \( T(n) \) step-by-step using the given recurrence relation.

### Step-by-Step Calculation

1. **Compute \( T(1) \):**

   Given:
   \[
   T(n) = 3T(n-1) + 12n
   \]
   Substitute \( n = 1 \):
   \[
   T(1) = 3T(0) + 12 \cdot 1
   \]
   We know \( T(0) = 5 \):
   \[
   T(1) = 3 \cdot 5 + 12 \cdot 1
   \]
   \[
   T(1) = 15 + 12
   \]
   \[
   T(1) = 27
   \]

2. **Compute \( T(2) \):**

   Substitute \( n = 2 \) into the recurrence relation:
   \[
   T(2) = 3T(1) + 12 \cdot 2
   \]
   We found \( T(1) = 27 \):
   \[
   T(2) = 3 \cdot 27 + 12 \cdot 2
   \]
   \[
   T(2) = 81 + 24
   \]
   \[
   T(2) = 105
   \]

### Conclusion

The value of t(2) is 105

To solve recurrence relations using the substitution method, we'll assume a solution form and then use induction to verify it. Let's tackle each recurrence relation one by one.

### a. \( T(n) = T(n-1) + c \)

**1. Assume a solution form:**

Assume \( T(n) = an + b \), where \( a \) and \( b \) are constants to be determined.

**2. Substitute into the recurrence relation:**

Substitute \( T(n) = an + b \) into \( T(n) = T(n-1) + c \):
\[
T(n) = a(n-1) + b + c
\]
\[
T(n) = an - a + b + c
\]

For this to match \( T(n) = an + b \), we must have:
\[
-an + a + b + c = b
\]
\[
-a + c = 0 \implies a = c
\]

Thus, the solution is:
\[
T(n) = cn + b
\]

**3. Determine \( b \) using initial conditions:**

If \( T(0) = T_0 \), then:
\[
T(0) = c \cdot 0 + b \implies b = T_0
\]

So the solution is:
\[
T(n) = cn + T_0
\]

### b. \( T(n) = 2T(n/2) + n \)

**1. Assume a solution form:**

Assume \( T(n) = an \log n + b \), where \( a \) and \( b \) are constants to be determined.

**2. Substitute into the recurrence relation:**

Substitute \( T(n) = an \log n + b \) into \( T(n) = 2T(n/2) + n \):
\[
T(n) = 2[a(n/2) \log (n/2) + b] + n
\]
\[
T(n) = 2\left[a(n/2) \log n - a(n/2) \log 2 + b\right] + n
\]
\[
T(n) = an \log n - a n + 2b + n
\]
\[
T(n) = an \log n + (n - a n + 2b)
\]

For this to match \( T(n) = an \log n + b \), we need:
\[
-n + a n = 0 \implies a = 1
\]
\[
2b = b \implies b = 2b
\]

Thus, the solution is:
\[
T(n) = n \log n + O(n)
\]

**In Big-O notation, the solution is \( T(n) = O(n \log n) \).**

### c. \( T(n) = 2T(n/2) + c \)

**1. Assume a solution form:**

Assume \( T(n) = a n + b \), where \( a \) and \( b \) are constants to be determined.

**2. Substitute into the recurrence relation:**

Substitute \( T(n) = a n + b \) into \( T(n) = 2T(n/2) + c \):
\[
T(n) = 2[a(n/2) + b] + c
\]
\[
T(n) = an + 2b + c
\]

For this to match \( T(n) = an + b \), we must have:
\[
2b + c = b \implies b = -c
\]

Thus, the solution is:
\[
T(n) = a n - c
\]

**Since \( a = c \), the solution is:**
\[
T(n) = c n - c
\]

**In Big-O notation, the solution is \( T(n) = O(n) \).**

### d. \( T(n) = T(n/2) + c \)

**1. Assume a solution form:**

Assume \( T(n) = a \log n + b \), where \( a \) and \( b \) are constants to be determined.

**2. Substitute into the recurrence relation:**

Substitute \( T(n) = a \log n + b \) into \( T(n) = T(n/2) + c \):
\[
T(n) = a \log (n/2) + b + c
\]
\[
T(n) = a (\log n - \log 2) + b + c
\]
\[
T(n) = a \log n - a \log 2 + b + c
\]

For this to match \( T(n) = a \log n + b \), we need:
\[
-a \log 2 + c = 0 \implies a = c
\]

Thus, the solution is:
\[
T(n) = c \log n + b - c \log 2
\]

**In Big-O notation, the solution is \( T(n) = O(\log n) \).**

ques3. Given a rec%rrence relation, solve it %sing the rec%rsive tree approach:
a. T(n) = 2T(n-1) +1
b. T(n) = 2T(n/2) + n



To solve recurrence relations using the recursive tree approach, we break down the problem into a tree structure where each node represents a subproblem. We then sum up the costs at each level of the tree to get the total cost of the problem. Let's tackle each recurrence relation using this approach.

### a. \( T(n) = 2T(n-1) + 1 \)

**1. ** Draw the Recursive Tree:**

- **Level 0 (Root):** The root of the tree is \( T(n) \).
  \[
  T(n)
  \]

- **Level 1:** The root splits into two subproblems:
  \[
  2T(n-1) + 1
  \]

- **Level 2:** Each of the two subproblems from Level 1 further splits into two:
  \[
  4T(n-2) + 2 \cdot 1 = 4T(n-2) + 2
  \]

- **Level 3:** Each of the four subproblems from Level 2 further splits into two:
  \[
  8T(n-3) + 4 \cdot 1 = 8T(n-3) + 4
  \]

**2. ** Determine the Number of Nodes and Cost at Each Level:**

- **Number of Nodes at Level \( i \):** \( 2^i \)
- **Cost at Each Node at Level \( i \):** Each node incurs a cost of 1 at that level.
  
  So, the total cost at level \( i \) is:
  \[
  \text{Cost at level } i = 2^i \cdot 1 = 2^i
  \]

**3. ** Determine the Number of Levels:**

- The recursion stops when \( n - i = 0 \), so the number of levels \( i \) is \( n \).

**4. ** Sum Up the Costs Across All Levels:**

The total cost \( T(n) \) is the sum of costs at all levels:
\[
T(n) = \sum_{i=0}^{n-1} 2^i
\]
This is a geometric series sum:
\[
T(n) = 2^n - 1
\]

**5. ** Final Solution:**

The recurrence relation \( T(n) = 2T(n-1) + 1 \) solves to:
\[
T(n) = 2^n - 1
\]

### b. \( T(n) = 2T(n/2) + n \)

**1. ** Draw the Recursive Tree:**

- **Level 0 (Root):** The root of the tree is \( T(n) \).
  \[
  T(n)
  \]

- **Level 1:** The root splits into two subproblems:
  \[
  2T(n/2) + n
  \]

- **Level 2:** Each of the two subproblems splits into two more subproblems:
  \[
  4T(n/4) + 2 \cdot (n/2) = 4T(n/4) + n
  \]

- **Level 3:** Each of the four subproblems splits into two:
  \[
  8T(n/8) + 4 \cdot (n/4) = 8T(n/8) + n
  \]

**2. ** Determine the Number of Nodes and Cost at Each Level:**

- **Number of Nodes at Level \( i \):** \( 2^i \)
- **Cost at Each Node at Level \( i \):** Each node incurs a cost of \( n / 2^i \) at that level.
  
  So, the total cost at level \( i \) is:
  \[
  \text{Cost at level } i = 2^i \cdot \frac{n}{2^i} = n
  \]

**3. ** Determine the Number of Levels:**

- The recursion stops when \( n / 2^i = 1 \), so the number of levels \( i \) is \( \log_2(n) \).

**4. ** Sum Up the Costs Across All Levels:**

The total cost \( T(n) \) is the sum of costs at all levels:
\[
T(n) = \sum_{i=0}^{\log_2(n)} n
\]
There are \( \log_2(n) + 1 \) levels, each contributing \( n \):
\[
T(n) = (\log_2(n) + 1) \cdot n
\]
\[
T(n) = n \log_2(n) + n
\]

**5. ** Final Solution:**

The recurrence relation \( T(n) = 2T(n/2) + n \) solves to:
\[
T(n) = O(n \log n)
\]

### Summary of Solutions:
- **a.** \( T(n) = 2^n - 1 \)
- **b.** \( T(n) = O(n \log n) \)
